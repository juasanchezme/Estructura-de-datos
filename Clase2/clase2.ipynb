```python
n = int(input()) #numero de lineas
listaNumeros = []
for i in range(n):
    entrada = int(input())
    listaNumeros.append(entrada)

sumaPositivo = 0
sumaNegativo = 0
for i in listaNumeros:
    if i>=0:
        sumaPositivo += i
    else:
        sumaNegativo += i

print(sumaPositivo, sumaNegativo)

```

```python
n = int(input()) #numero de lineas

numeros = [int(input()) for i in range(n)]
sPositiva = sum([i for i in numeros if i>= 0])
sNegativa = sum([i for i in numeros if i< 0])
print(sPositiva, sNegativa)
```
