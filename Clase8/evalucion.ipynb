```python
import heapq

# Inicializamos el montículo
personas = []  # Usaremos un heap de máximos (con valores negativos)

while True:
    entrada = input().strip()
    
    if entrada == "fin":
        break

    comando = entrada.split()
    solicitud = comando[0]

    if solicitud == "ingresa":
        prestigio = int(comando[1])
        
        if personas:
            # Obtener el mayor prestigio (negativo, ya que usamos heap mínimo invertido)
            mayorPrestigio = -personas[0]
            
            # Verificar si puede entrar (debe tener al menos la mitad del mayor)
            if prestigio >= mayorPrestigio/2:
                heapq.heappush(personas, -prestigio)  # Agregar con signo negativo
                print("adelante")
            else:
                print("denegado")
        else:
            # Si el club está vacío, cualquier persona puede entrar
            heapq.heappush(personas, -prestigio)
            print("adelante")

    elif solicitud == "salida":
        if personas:
            # Sacar al de mayor prestigio
            heapq.heappop(personas)
            print("hasta pronto")
```
