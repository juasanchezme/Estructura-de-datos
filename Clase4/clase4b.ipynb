**Solucin alternativa:**
En el contexto de este problema, el mdulo de un nmero impar siempre ser 1 cuando se calcula con respecto a 2.
Por ejemplo:  
 4%2=0
Si un nmero es impar, su mdulo con respecto a 2 es 1 (por ejemplo, 
3%2=1).

```python
# Lectura de entrada
N = int(input())
so = list(map(int, input().split(", ")))
lar = list(map(int, input().split(", ")))
isz = list(map(int, input().split(", ")))

# Inicializacin de puntajes
puntajeSo, puntajeLar, puntajeIsz = 0, 0, 0

# Iteracin directa sobre los turnos
for i in range(N):
    suma = so[i] + lar[i] + isz[i]  # Suma del turno actual
    
    # Evaluar para cada jugador
    if (so[i] % 2 == suma % 2):  # Paridad coincide
        puntajeSo += 1
    if (lar[i] % 2 == suma % 2):
        puntajeLar += 1
    if (isz[i] % 2 == suma % 2):
        puntajeIsz += 1

# Salida
print(f"SO:{puntajeSo}, LAR:{puntajeLar}, IS:{puntajeIsz}")
```

## Ejercicio 3

**Clave para solucionar el ejercicio** 

Python permite realizar el **intercambio de dos valores en una sola lnea** gracias a la desempaquetacin de tuplas:

a, b = b, a

Otra linea inportante es la que nos permite **invertir los elementos de una lista**

caracteres = caracteres[::-1]


```python
def desencriptar_mensaje(mensaje):
    # Eliminar espacios y convertir a lista 
    caracteres = list(mensaje.split(" "))
    
    # Paso 1: Esta linea invierte el orden de la lista
    caracteres = caracteres[::-1]
    
    # Paso 2: Deshacer el intercambio de caracteres en posiciones impares
    # Creamos el for para que solo recorra las posiciones impares, empieza en 1 y va en pasos de 2; 1, 3, 5, 7,...
    for i in range(1, len(caracteres), 2):
        caracteres[i], caracteres[i - 1] = caracteres[i - 1], caracteres[i]
    
    # Convertir de nuevo a string y retornarlo
    return "".join(caracteres)

# Leer la cantidad de casos de prueba
C = int(input())
resultados = []

# Procesar cada caso de prueba
for _ in range(C):
    mensaje_encriptado = input()
    resultado = desencriptar_mensaje(mensaje_encriptado)
    resultados.append(resultado)

# Imprimir los resultados
for resultado in resultados:
    print(resultado)
```

## Ejercicio 4

Cuando tenemos un arreglo con N elementos y necesitamos recorrerlo por alguna razn, lo ms natural es hacerlo de izquierda a derecha, verdad? Esto es, desde el ndice 0 hasta el N-1 aumentando de a uno.

Bueno, pues en un recorrido arcoris tambin se pasa por todos los elementos una nica vez, pero en un orden bastante peculiar. Como se muestra en la figura, se parte del ndice 0, de all se va para el N-1, de all para el 1, de all se va para el N-2, de all para el 2, de all se va para el N-3, y as sucesivamente hasta llegar al "centro" del arreglo.

```python
import math
N = int(input())
entrada = list(input().split(", "))

salida =[]
for i in range(math.floor(N/2)):
    salida.append(entrada[i])
    salida.append(entrada[(N-1)-i])
    
# Si el nmero de elementos en la lista es impar, aade el elemento del medio
if N % 2 != 0:
    salida.append(entrada[math.floor(N / 2)])

print("".join(salida))
```

En este caso importamos la libreria Math, nativa de python para usar el mtodo .floor() que redondea el flotante al entero por debajo, y tener en cuenta que en una lista impar nos toca aadir el elemento del medio ya que en el ciclo for no se estara teniendo en cuenta

### Versin alternativa

repartir elementos a cada integrante de una lista

```python
def distribuir_elementos(lista, n_integrantes):
    # Crear una lista vaca para cada integrante
    participantes = [[] for _ in range(n_integrantes)]
    
    # Asignar elementos uno por uno
    for i, elemento in enumerate(lista):
        participantes[i % n_integrantes].append(elemento)
    
    return participantes

# Ejemplo de uso
elementos = ["E1", "E2", "E3", "E4", "E5", "E6", "E7"]
n_integrantes = 3
resultado = distribuir_elementos(elementos, n_integrantes)

# Mostrar resultado
for i, integrante in enumerate(resultado):
    print(f"Integrante {chr(65+i)}: {integrante}")
```
