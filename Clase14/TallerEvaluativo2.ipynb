```python
gameOver = False
jugadores = dict()

def caparazon():
    if not jugadores:  # Si no hay jugadores
        return "SIN OBJETIVO"
        
    # Encontrar la posición máxima
    max_pos = max(jugadores.values())
    
    # Contar cuántos jugadores están en esa posición
    lideres = [id for id, pos in jugadores.items() if pos == max_pos]
    
    # Si hay más de un líder, no hay objetivo
    if len(lideres) > 1:
        return "SIN OBJETIVO"
    else:
        return lideres[0]

while not gameOver:
    comando = input().split()
    
    if comando[0] == "C.AZUL":
        print(caparazon())
    elif comando[0] == "0":  # Comparar con string "0", no con int
        gameOver = True
    else:
        id, posicion = comando[0], int(comando[1])
        
        # Actualizar posición del jugador
        if id not in jugadores:
            jugadores[id] = posicion
        else:
            jugadores[id] += posicion

```

```python
def solve():
    jugadores = {}            # Rastrea posición de cada jugador
    max_pos = 0               # Posición máxima actual
    lideres_count = 0         # Cuántos jugadores están en la posición máxima
    lider_actual = None       # ID del líder único (si existe)
    
    while True:
        entrada = input().strip()
        if entrada == "0":
            break
            
        if entrada == "C.AZUL":
            # Responder consulta
            if lideres_count != 1:
                print("SIN OBJETIVO")
            else:
                print(lider_actual)
        else:
            # Procesar movimiento
            id_jugador, avance = entrada.split()
            avance = int(avance)
            
            # Actualizar posición
            pos_anterior = jugadores.get(id_jugador, 0)
            nueva_pos = pos_anterior + avance
            jugadores[id_jugador] = nueva_pos
            
            # Actualizar información de liderazgo
            if nueva_pos > max_pos:
                # Nuevo líder único
                max_pos = nueva_pos
                lideres_count = 1
                lider_actual = id_jugador
            elif nueva_pos == max_pos:
                # Nuevo empate en primera posición
                lideres_count += 1
                lider_actual = None  # Ya no hay líder único
            elif pos_anterior == max_pos:
                # Este jugador era líder pero ya no
                lideres_count -= 1
                
                # Si ahora queda un solo líder, necesitamos encontrarlo
                if lideres_count == 1:
                    for jug, pos in jugadores.items():
                        if pos == max_pos:
                            lider_actual = jug
                            break


solve()
```

```python
import sys
from heapq import heappush, heappop

def main():
    input = sys.stdin.read().split()
    idx = 0
    player_pos = {}
    pos_players = {}
    heap = []
    
    while idx < len(input):
        line = input[idx]
        if line == '0':
            break
        if line == 'C.AZUL':
            current_max = 0
            while heap:
                candidate = -heap[0]
                if candidate in pos_players and len(pos_players[candidate]) > 0:
                    current_max = candidate
                    break
                else:
                    heappop(heap)
            if current_max == 0:
                print("SIN OBJETIVO")
            else:
                players = pos_players[current_max]
                if len(players) == 1:
                    print(next(iter(players)))
                else:
                    print("SIN OBJETIVO")
            idx += 1
        else:
            X = int(line)
            Y = int(input[idx + 1])
            idx += 2
            old_p = player_pos.get(X, 0)
            new_p = old_p + Y
            player_pos[X] = new_p
            
            if old_p in pos_players:
                if X in pos_players[old_p]:
                    pos_players[old_p].remove(X)
                    if not pos_players[old_p]:
                        del pos_players[old_p]
            
            if new_p not in pos_players:
                pos_players[new_p] = set()
            pos_players[new_p].add(X)
            heappush(heap, -new_p)


main()
```
