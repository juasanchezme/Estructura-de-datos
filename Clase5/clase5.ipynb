### Correccin


```python
listaNumeros = []  

while True:
    entrada = input().strip()  
    
    if entrada == "E":  # Comando para terminar el programa
        break
    
    letra, numero = entrada.split()  # Dividir en letra y nmero
    numero = int(numero)  
    
    if letra == "A": 
        listaNumeros.append(numero)
    
    elif letra == "M":  # Calcular suma de mltiplos
        sumaNumero = sum(i for i in listaNumeros if i % numero == 0)
        print(sumaNumero)  
```
