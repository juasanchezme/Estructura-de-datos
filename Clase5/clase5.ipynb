```python
def juego_sorpresa(n, k):
    estudiantes = list(range(1, n + 1))  # Lista de estudiantes numerados del 1 al N
    posicion = 0  # Posicin inicial en el crculo

    while len(estudiantes) > 1:
        # Calcular la posicin del estudiante que va a salir
        posicion = (posicion + k - 1) % len(estudiantes)
        estudiante_saliente = estudiantes[posicion]
        
        # Eliminar al estudiante de la lista
        estudiantes.pop(posicion)
        
        # Actualizar K como mdulo del estudiante que sali y los restantes
        if len(estudiantes) > 0:  # Evitar divisin por cero
            k = estudiante_saliente % len(estudiantes)
            if k == 0:
                k = 1  # Si el mdulo es 0, se toma como 1

    # Retornar el nico estudiante restante
    return estudiantes[0]


# Lectura de entrada
n_casos = int(input())
resultados = []

for _ in range(n_casos):
    n, k = map(int, input().split())
    resultados.append(juego_sorpresa(n, k))

# Imprimir resultados
for resultado in resultados:
    print(resultado)
```

```python
jugadores = []

gameOver = False

while not gameOver:
    k = int(input())

    if k not in jugadores:
        jugadores.append(k)

    if k+1 in jugadores:
        jugadores.remove(k+1)
        jugadores.remove(k)

    if k-1 in jugadores:
        jugadores.remove(k-1)
        try:
            jugadores.remove(k)
        except:
            continue

    if k == 0:
        gameOver = True

print(jugadores)
```

```python
# Lista para almacenar los jugadores sobrevivientes (en el orden en que entran)
jugadores = []

# Conjunto para verificar los nmeros que ya estn en la habitacin (para evitar bsquedas innecesarias)
numeros_en_habitacion = set()

while True:
    # Leemos la entrada
    numero = int(input().strip())
    
    # Si el nmero es 0, terminamos la entrada
    if numero == 0:
        break
    
    # Verificamos si el nmero ya est en la habitacin
    if numero in numeros_en_habitacion:
        # Si ya est, eliminamos ambos jugadores (el actual y el que ya estaba)
        jugadores = [jug for jug in jugadores if jug != numero]
        numeros_en_habitacion.remove(numero)
    else:
        # Verificamos si hay jugadores con nmero vecino (k+1 o k-1)
        if numero - 1 in numeros_en_habitacion:
            # Si hay un jugador con k-1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero - 1]
            numeros_en_habitacion.remove(numero - 1)
        elif numero + 1 in numeros_en_habitacion:
            # Si hay un jugador con k+1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero + 1]
            numeros_en_habitacion.remove(numero + 1)
        else:
            # Si no hay eliminaciones, agregamos el jugador a la lista de sobrevivientes
            jugadores.append(numero)
            numeros_en_habitacion.add(numero)

# Si quedan jugadores, los mostramos
if jugadores:
    print(" ".join(map(str, jugadores)))
else:
    # Si no quedan jugadores, mostramos 0
    print(0)
```

```python
# Lista para almacenar los nmeros de los jugadores
jugadores = []
# Conjunto para verificar los nmeros en uso
nmeros_en_uso = set()

while True:
    # Leemos la entrada
    numero = int(input().strip())
    
    # Si el nmero es 0, terminamos la entrada
    if numero == 0:
        break
    
    # Verificamos si el nmero ya ha sido elegido antes
    if numero in nmeros_en_uso:
        # Si ya ha sido elegido, eliminamos este nmero de los jugadores
        jugadores = [jug for jug in jugadores if jug != numero]
    else:
        # Verificamos si hay algn jugador con k-1 o k+1
        if (numero - 1) in nmeros_en_uso:
            # Si hay un jugador con k-1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero and jug != (numero - 1)]
            nmeros_en_uso.discard(numero - 1)  # Eliminamos el jugador con k-1
        elif (numero + 1) in nmeros_en_uso:
            # Si hay un jugador con k+1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero and jug != (numero + 1)]
            nmeros_en_uso.discard(numero + 1)  # Eliminamos el jugador con k+1
        else:
            # Si no hay conflictos, agregamos el nmero a la lista de jugadores
            jugadores.append(numero)
    
    # Aadimos el nmero al conjunto para evitar duplicados
    nmeros_en_uso.add(numero)

# Si quedan jugadores, los mostramos
if jugadores:
    print(" ".join(map(str, jugadores)))
else:
    # Si no quedan jugadores, mostramos 0
    print(0)
```

## Ejercicio 5

```python
import math

def insertar_ordenado(lista, num):
    """Inserta un nmero en una lista de manera ordenada."""
    left, right = 0, len(lista)
    while left < right:
        mid = (left + right) // 2
        if lista[mid] < num:
            left = mid + 1
        else:
            right = mid
    lista.insert(left, num)

def calcular_mediana(lista):
    """Calcula la mediana de una lista ordenada."""
    n = len(lista)
    if n % 2 == 1:
        return lista[n // 2]
    else:
        # Promedio de los dos valores centrales
        num1 = lista[n // 2 - 1]
        num2 = lista[n // 2]
        # Calculamos la fraccin (num1 + num2) / 2
        numerador = num1 + num2
        denominador = 2
        # Simplificamos la fraccin
        gcd_val = math.gcd(numerador, denominador)
        numerador //= gcd_val
        denominador //= gcd_val
        if denominador == 1:
            return numerador  # Si la fraccin es un nmero entero
        return f"{numerador}/{denominador}"

def procesar_mediana(M, numeros):
    lista = []
    result = []
    
    for i, num in enumerate(numeros):
        insertar_ordenado(lista, num)
        
        # Si el tamao de la lista es mltiplo de M, calculamos la mediana
        if (i + 1) % M == 0:
            mediana = calcular_mediana(lista)
            result.append(str(mediana))
    
    return result

# Leer la entrada
M = int(input())
numeros = []
while True:
    num = int(input())
    if num == 0:
        break
    numeros.append(num)

# Procesamos la mediana
resultados = procesar_mediana(M, numeros)

# Imprimimos los resultados
for res in resultados:
    print(res)
```

## Ejercicio 6

```python
# Lista para almacenar la fila de fanticos
fila = []

while True:
    # Leemos la entrada
    identificacion, maximo = input().split()
    maximo = int(maximo)
    
    # Si encontramos la lnea de "0 0", terminamos
    if identificacion == "0" and maximo == 0:
        break
    
    # Si el tamao de la fila es menor que el mximo del fantico, se forma
    if len(fila) < maximo:
        fila.append((identificacion, maximo))
    else:
        # Si el tamao de la fila es mayor o igual que el mximo, buscar al que se debe expulsar
        # Filtramos el fantico con el mnimo "maximo" que se uni primero
        expulsar = min(fila, key=lambda x: x[1] if x[1] <= maximo else float('inf'))
        fila.remove(expulsar)
        fila.append((identificacion, maximo))

# Imprimimos el tamao final de la fila
print(len(fila))
```
