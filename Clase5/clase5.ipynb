# Arreglos de redimensionamiento din치mico

son estructuras de datos que permiten ajustar su tama침o en tiempo de ejecuci칩n para acomodar elementos adicionales sin necesidad de especificar un tama침o fijo al momento de crearlos. 

## Inserci칩n 

En un arreglo din치mico solo se aumenta la capacidad
cuando, al momento de hacer una inserci칩n, no hay
suficiente espacio. 
ejemplo donde vamos
a insertar uno a uno, y al final del arreglo, N=9
elementos partiendo de un arreglo inicialmente vac칤o
de capacidad 2 y que el factor de crecimiento es 2.0

La clase list en Python act칰a como un arreglo din치mico. Ejemplo:

```python
arr = [1, 2, 3]
arr.append(4)  # Agrega un nuevo elemento
print(arr)  
```

### Acceso a elementos



```python

mi_lista = [10, 20, 30, 40]

# Acceso por 칤ndice
print(mi_lista[0])  # 10

# 칈ndices negativos
print(mi_lista[-1])  # 40 (칰ltimo elemento)

# Acceso a un rango (slicing)
print(mi_lista[1:3])  # [20, 30]
```

### modificar

```python
mi_lista = [1, 2, 3]
mi_lista[1] = 20  # Modifica el segundo elemento
print(mi_lista)  # [1, 20, 3]
```

### Agregar

```python
mi_lista = [1, 2, 3]

# Agregar al final
mi_lista.append(4)
print(mi_lista)  # [1, 2, 3, 4]

# Insertar en una posici칩n espec칤fica
mi_lista.insert(1, 15)  # Inserta 15 en la posici칩n 1
print(mi_lista)  # [1, 15, 2, 3, 4]

# Extender con otra lista
mi_lista.extend([5, 6])
print(mi_lista)  # [1, 15, 2, 3, 4, 5, 6]
```

### Eliminar

```python
mi_lista = [1, 2, 3, 4, 5]

# Eliminar por valor
mi_lista.remove(3)  # Elimina el primer 3 que encuentra
print(mi_lista)  # [1, 2, 4, 5]

# Eliminar por 칤ndice
del mi_lista[1]  # Elimina el elemento en la posici칩n 1
print(mi_lista)  # [1, 4, 5]

# Eliminar y obtener el 칰ltimo elemento
ultimo = mi_lista.pop()
print(ultimo)  # 5
print(mi_lista)  # [1, 4]

lista2 = list(range(0, 10))
print(lista2)
# Eliminar y obtener el elemento de un indice
ultimo2 = lista2.pop(1)
print(ultimo2)  # 5
print(lista2)  # [1, 4]

del lista2[-3:]  # Eliminar los 칰ltimos k elementos
print(lista2)

# Eliminar todos los elementos
mi_lista.clear()
print(mi_lista)  # []
```

### Ordenar y revertir

```python
mi_lista = [5, 3, 8, 1]

# Ordenar ascendente
mi_lista.sort()
print(mi_lista)  # [1, 3, 5, 8]

# Ordenar descendente
mi_lista.sort(reverse=True)
print(mi_lista)  # [8, 5, 3, 1]

# Revertir el orden actual
mi_lista.reverse()
print(mi_lista)  # [1, 3, 5, 8]
```

### List comprehension

```python
# Crear una lista con cuadrados de n칰meros
cuadrados = [x**2 for x in range(5)]
print(cuadrados)  # [0, 1, 4, 9, 16]

# Filtrar elementos
pares = [x for x in range(10) if x % 2 == 0]
print(pares)  # [0, 2, 4, 6, 8]
```

```python
def juego_sorpresa(n, k):
    estudiantes = list(range(1, n + 1))  # Lista de estudiantes numerados del 1 al N
    posicion = 0  # Posici칩n inicial en el c칤rculo

    while len(estudiantes) > 1:
        # Calcular la posici칩n del estudiante que va a salir
        posicion = (posicion + k - 1) % len(estudiantes)
        estudiante_saliente = estudiantes[posicion]
        
        # Eliminar al estudiante de la lista
        estudiantes.pop(posicion)
        
        # Actualizar K como m칩dulo del estudiante que sali칩 y los restantes
        if len(estudiantes) > 0:  # Evitar divisi칩n por cero
            k = estudiante_saliente % len(estudiantes)
            if k == 0:
                k = 1  # Si el m칩dulo es 0, se toma como 1

    # Retornar el 칰nico estudiante restante
    return estudiantes[0]


# Lectura de entrada
n_casos = int(input())
resultados = []

for _ in range(n_casos):
    n, k = map(int, input().split())
    resultados.append(juego_sorpresa(n, k))

# Imprimir resultados
for resultado in resultados:
    print(resultado)
```

```python
jugadores = []

gameOver = False

while not gameOver:
    k = int(input())

    if k not in jugadores:
        jugadores.append(k)

    if k+1 in jugadores:
        jugadores.remove(k+1)
        jugadores.remove(k)

    if k-1 in jugadores:
        jugadores.remove(k-1)
        try:
            jugadores.remove(k)
        except:
            continue

    if k == 0:
        gameOver = True

print(jugadores)
```

```python
# Lista para almacenar los jugadores sobrevivientes (en el orden en que entran)
jugadores = []

# Conjunto para verificar los n칰meros que ya est치n en la habitaci칩n (para evitar b칰squedas innecesarias)
numeros_en_habitacion = set()

while True:
    # Leemos la entrada
    numero = int(input().strip())
    
    # Si el n칰mero es 0, terminamos la entrada
    if numero == 0:
        break
    
    # Verificamos si el n칰mero ya est치 en la habitaci칩n
    if numero in numeros_en_habitacion:
        # Si ya est치, eliminamos ambos jugadores (el actual y el que ya estaba)
        jugadores = [jug for jug in jugadores if jug != numero]
        numeros_en_habitacion.remove(numero)
    else:
        # Verificamos si hay jugadores con n칰mero vecino (k+1 o k-1)
        if numero - 1 in numeros_en_habitacion:
            # Si hay un jugador con k-1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero - 1]
            numeros_en_habitacion.remove(numero - 1)
        elif numero + 1 in numeros_en_habitacion:
            # Si hay un jugador con k+1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero + 1]
            numeros_en_habitacion.remove(numero + 1)
        else:
            # Si no hay eliminaciones, agregamos el jugador a la lista de sobrevivientes
            jugadores.append(numero)
            numeros_en_habitacion.add(numero)

# Si quedan jugadores, los mostramos
if jugadores:
    print(" ".join(map(str, jugadores)))
else:
    # Si no quedan jugadores, mostramos 0
    print(0)
```

```python
# Lista para almacenar los n칰meros de los jugadores
jugadores = []
# Conjunto para verificar los n칰meros en uso
n칰meros_en_uso = set()

while True:
    # Leemos la entrada
    numero = int(input().strip())
    
    # Si el n칰mero es 0, terminamos la entrada
    if numero == 0:
        break
    
    # Verificamos si el n칰mero ya ha sido elegido antes
    if numero in n칰meros_en_uso:
        # Si ya ha sido elegido, eliminamos este n칰mero de los jugadores
        jugadores = [jug for jug in jugadores if jug != numero]
    else:
        # Verificamos si hay alg칰n jugador con k-1 o k+1
        if (numero - 1) in n칰meros_en_uso:
            # Si hay un jugador con k-1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero and jug != (numero - 1)]
            n칰meros_en_uso.discard(numero - 1)  # Eliminamos el jugador con k-1
        elif (numero + 1) in n칰meros_en_uso:
            # Si hay un jugador con k+1, eliminamos ambos
            jugadores = [jug for jug in jugadores if jug != numero and jug != (numero + 1)]
            n칰meros_en_uso.discard(numero + 1)  # Eliminamos el jugador con k+1
        else:
            # Si no hay conflictos, agregamos el n칰mero a la lista de jugadores
            jugadores.append(numero)
    
    # A침adimos el n칰mero al conjunto para evitar duplicados
    n칰meros_en_uso.add(numero)

# Si quedan jugadores, los mostramos
if jugadores:
    print(" ".join(map(str, jugadores)))
else:
    # Si no quedan jugadores, mostramos 0
    print(0)
```

```python
import math

def insertar_ordenado(lista, num):
    """Inserta un n칰mero en una lista de manera ordenada."""
    left, right = 0, len(lista)
    while left < right:
        mid = (left + right) // 2
        if lista[mid] < num:
            left = mid + 1
        else:
            right = mid
    lista.insert(left, num)

def calcular_mediana(lista):
    """Calcula la mediana de una lista ordenada."""
    n = len(lista)
    if n % 2 == 1:
        return lista[n // 2]
    else:
        # Promedio de los dos valores centrales
        num1 = lista[n // 2 - 1]
        num2 = lista[n // 2]
        # Calculamos la fracci칩n (num1 + num2) / 2
        numerador = num1 + num2
        denominador = 2
        # Simplificamos la fracci칩n
        gcd_val = math.gcd(numerador, denominador)
        numerador //= gcd_val
        denominador //= gcd_val
        if denominador == 1:
            return numerador  # Si la fracci칩n es un n칰mero entero
        return f"{numerador}/{denominador}"

def procesar_mediana(M, numeros):
    lista = []
    result = []
    
    for i, num in enumerate(numeros):
        insertar_ordenado(lista, num)
        
        # Si el tama침o de la lista es m칰ltiplo de M, calculamos la mediana
        if (i + 1) % M == 0:
            mediana = calcular_mediana(lista)
            result.append(str(mediana))
    
    return result

# Leer la entrada
M = int(input())
numeros = []
while True:
    num = int(input())
    if num == 0:
        break
    numeros.append(num)

# Procesamos la mediana
resultados = procesar_mediana(M, numeros)

# Imprimimos los resultados
for res in resultados:
    print(res)
```

```python
# Lista para almacenar la fila de fan치ticos
fila = []

while True:
    # Leemos la entrada
    identificacion, maximo = input().split()
    maximo = int(maximo)
    
    # Si encontramos la l칤nea de "0 0", terminamos
    if identificacion == "0" and maximo == 0:
        break
    
    # Si el tama침o de la fila es menor que el m치ximo del fan치tico, se forma
    if len(fila) < maximo:
        fila.append((identificacion, maximo))
    else:
        # Si el tama침o de la fila es mayor o igual que el m치ximo, buscar al que se debe expulsar
        # Filtramos el fan치tico con el m칤nimo "maximo" que se uni칩 primero
        expulsar = min(fila, key=lambda x: x[1] if x[1] <= maximo else float('inf'))
        fila.remove(expulsar)
        fila.append((identificacion, maximo))

# Imprimimos el tama침o final de la fila
print(len(fila))
```

```python
# Definimos la fila como una lista vac칤a
fila = []

# Leemos la entrada
while True:
    # Obtenemos la entrada del usuario
    entrada = input().strip()
    
    # Si la entrada es "0 0", terminamos
    if entrada == "0 0":
        break
    
    # Separamos el ID y el tama침o m치ximo
    id_persona, max_tamano = entrada.split()
    max_tamano = int(max_tamano)
    
    # Mientras la fila est칠 llena y el tama침o m치ximo del primero de la fila sea menor que el actual
    while len(fila) > 0 and len(fila) > max_tamano:
        # Si el tama침o de la fila es mayor que el m치ximo del primer elemento, lo eliminamos
        fila.pop(0)
    
    # Si la longitud de la fila es menor o igual al m치ximo, agregamos el fan치tico
    fila.append((id_persona, max_tamano))

# El tama침o final de la fila es el n칰mero de elementos que quedan en la lista
print(len(fila))
```
