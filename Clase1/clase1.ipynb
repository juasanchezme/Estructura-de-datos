# 📚 ¿Qué es una Estructura de Datos?

Una **estructura de datos** es una manera de almacenar y organizar una colección de datos en la computadora para que estos puedan ser manipulados de manera eficiente dentro de un determinado algoritmo. 💻

## 🔹 ¿Qué son los Datos?
los **datos** son cualquier tipo de información que se puede almacenar, manipular y transmitir, tenemos 2 tipos de datos

### 1. 🧩 Datos Primitivos
Los datos primitivos que son atómicos, lo que significa que no se pueden descomponer en otros tipos de datos más pequeños. Algunos ejemplos comunes son:

- Entero (int)
- Flotante (float)
- Caracter (char)
- Booleano (bool)

### 2. 🧳 Datos Compuestos
Los datos compuestos son estructuras de datos que pueden almacenar múltiples valores, que pueden ser del mismo tipo o de tipos distintos. Pueden estar formados por datos primitivos o incluso por otros datos compuestos. Algunos ejemplos son:

- Arreglos o listas (arrays/lists): 
- Estructuras o registros (structs): agrupaciones de datos de diferentes tipos bajo un solo nombre, como una estructura Persona con nombre, edad, y altura.
- Cadenas de texto (strings); dado que  son secuencias de caracteres.
- Objetos: en programación orientada a objetos, combinan datos (atributos) y métodos para manipularlos.

# Eficiencia algorítmica y notación Big O

Considerando como recursos el procesador y la memoria, podemos medir:
**f**: Cantidad de operaciones realizadas
**g** Bytes utilizados durante la ejecución

## Problema:
Leer un valor entero positivo N y mostrar la sumatoria
de los enteros entre 1 y N (incluyéndolos)

```python
# Solución 1

N = int(input("ingrese el valor de N "))
numeros = []
for i in range(0, N+1):
    numeros.append(i)

sumaNumeros = 0
for i in numeros:
    sumaNumeros += i

print(sumaNumeros) 
```

```python
# Solución 2

N = int(input("ingrese el valor de N "))
nums = list(range(1, N+1)) 
print(sum(nums))
```

***Operaciones:*** f(N) =  1 + N + N = 2N + 1 → 𝑂(𝑁)

***Memoria:*** g(N) = 1 + N → 𝑂(𝑁)

```python
# Solución 3

N = int(input("ingrese el valor de N "))

nuevaSuma = 0
for i in range(1, N+1):
    nuevaSuma += i

print(nuevaSuma)
```
